```python
'''
Q2. What is a catter plot? Use the following code to generate data for x and y . Using this generated data plot a scatter plot.
Ans - A scatter plot is a diagram where each value in the data set is represented by a dot.
'''
# code-
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Graph')
plt.show()
```

```python
'''
Q3. WHy is the subplot() function used? Draw four line plots using the subplot() function.
Ans - Subplots method provides a way to plot multiple plots on a single figure
'''
```

```python
fig,a = plt.subplots(2,2)
x = np.array([0,1,2,3,4,5])
y = np.array([0,100,200,300,400,500])
a[0][0].plot(x,y)
a[0][0].set_title('For line1')

x = np.array([0,1,2,3,4,5])
y = np.array([50,20,40,20,60,70])
a[1][0].plot(x,y)
a[0][1].set_title('For Line2')

x = np.array([0,1,2,3,4,5])
y = np.array([10,20,30,40,50,60])
a[0][1].plot(x,y)
a[0][1].set_title('For Line2')

x = np.array([0,1,2,3,4,5])
y = np.array([200, 350, 250, 550, 450, 150])
a[1][1].plot(x,y)
a[1][1].set_title('For Line4')
plt.show()
```

```python
'''
 Q4. What is a bar plot? what is it used? Using the following data plot a bar piot and a horizintal bar plot.
 Ans- Bar plots ara a types of data visualization used to represnt data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the  category of the data. The Matplotlib library in Python is widely used to create bar plots.
'''
company = np.array(['Apple', 'Microsoft', 'Google', 'AMD'])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.show()
```

```python
'''
Q5. What is a box plot? Why is it used? Using the following data plot a box plot.
Ans- Box plots are used to show distribution of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high level information at a glance, offering generl iformation about a goup of data's symmerty, skew, variance, and outliers.
'''
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
dataset = [box1,box2]
figure = plt.figure(figsize=(10,7))
ax = figure.add_axes([0,0,1,1])
bp = ax.boxplot(dataset)
plt.show()
```
