Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

Ans- A t-test is used when the population standard deviation is unknown or when the sample size is small (usually less than 30). The test uses the t-distribution to calculate the probability of obtaining a difference in means as extreme as the one observed in the sample, assuming the null hypothesis is true. The degrees of freedom for the t-distribution depend on the sample size.

A z-test, on the other hand, is used when the population standard deviation is known or when the sample size is large (usually greater than 30). The test uses the standard normal distribution to calculate the probability of obtaining a difference in means as extreme as the one observed in the sample, assuming the null hypothesis is true.

Example scenario for t-test:

Suppose you want to test whether there is a difference in mean exam scores between two groups of students - one group who took a standardized test preparation course and another group who did not take the course. You randomly select 20 students from each group and calculate their average scores. Since the population standard deviation is unknown, you would use a two-sample t-test to determine if there is a significant difference in the means of the two groups.

Example scenario for z-test:

Suppose you want to test whether a new weight loss pill is effective at reducing body weight. You randomly select 200 participants and divide them into two groups - one group receives the weight loss pill, and the other group receives a placebo. You measure the weight of each participant before and after the intervention and calculate the difference in weight for each individual. Since the sample size is large, you can use a two-sample z-test to determine if there is a significant difference in the means of the two groups.

Q2: Differentiate between one-tailed and two-tailed tests.

A one-tailed test, also known as a directional test, is used when we have a specific hypothesis about the direction of the difference between two groups or variables. For example, if we hypothesize that Group A will have higher scores than Group B, we are using a one-tailed test. The one-tailed test is called directional because it tests for differences in only one direction, either positive or negative.

A two-tailed test, also known as a non-directional test, is used when we have no specific hypothesis about the direction of the difference between two groups or variables. For example, if we hypothesize that there is a difference between Group A and Group B, but we do not know whether Group A will have higher or lower scores than Group B, we are using a two-tailed test. The two-tailed test tests for differences in both directions, positive and negative.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Ans- Type 1 error occurs when we reject a null hypothesis that is actually true. It is also known as a false positive. The probability of making a Type 1 error is denoted by the symbol alpha (α) and is set by the researcher before conducting the test. The alpha level is typically set at 0.05 or 0.01.

An example scenario for Type 1 error is a medical test for a rare disease that has a false positive rate of 5%. This means that out of 100 people who are tested, 5 will receive a positive result even though they do not have the disease. This can lead to unnecessary anxiety, further testing, and treatments, which can be expensive and potentially harmful.

Type 2 error occurs when we fail to reject a null hypothesis that is actually false. It is also known as a false negative. The probability of making a Type 2 error is denoted by the symbol beta (β). The power of a test is equal to 1 - β and represents the probability of correctly rejecting a null hypothesis when it is actually false.

An example scenario for Type 2 error is a medical test for a disease that has a false negative rate of 10%. This means that out of 100 people who have the disease, 10 will receive a negative result and will not be diagnosed with the disease. This can lead to delayed treatment, which can have serious consequences for the patient's health.

The formula for Bayes's theorem is as follows:

P(A|B) = P(B|A) * P(A) / P(B)

An example scenario to understand Bayes's theorem is a medical test for a rare disease. Suppose the prevalence of the disease in the population is 1%, and the test has a sensitivity of 90% (the probability of testing positive given that the person has the disease) and a specificity of 95% (the probability of testing negative given that the person does not have the disease).

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.


Ans- To conduct a hypothesis test using a t-test, we need to state the null and alternative hypotheses:

Null hypothesis (H0): The mean weight loss of the population is equal to 0 (i.e., the drug is not effective).
Alternative hypothesis (Ha): The mean weight loss of the population is greater than 0 (i.e., the drug is effective).

We will use a one-sample t-test to test the hypotheses since we are comparing the sample mean to a known population mean of 0.

The test statistic can be calculated as:

t = (x̄ - μ) / (s / √n)

where:
x̄ is the sample mean
μ is the population mean (here, μ = 0)
s is the sample standard deviation
n is the sample size

Substituting the given values, we get:

t = (6 - 0) / (2.5 / √50) = 12.65

The degrees of freedom (df) for this test is n - 1 = 49.

Using a t-distribution table or a calculator, we can find the critical t-value for a one-tailed test with a 95% confidence level and 49 degrees of freedom, which is approximately 1.677.

Since the calculated t-value of 12.65 is greater than the critical t-value of 1.677, we reject the null hypothesis and conclude that the weight loss drug is significantly effective at a 95% confidence level.

Interpretation: The study provides sufficient evidence to suggest that the new weight loss drug is significantly effective in helping participants lose weight, with an average weight loss of 6 pounds in the sample.
